```python
a = int(input("첫 번째 숫자를 입력하세요: "))  # 첫 번째 정수 입력
b = int(input("두 번째 숫자를 입력하세요: "))  # 두 번째 정수 입력

print("a + b =", a + b)  # 덧셈
print("a - b =", a - b)  # 뺄셈
print("a * b =", a * b)  # 곱셈
print("a / b =", a / b)  # 나눗셈
```
